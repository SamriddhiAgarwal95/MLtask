```python
import pandas as pd
import numpy as np

# Load the file
df = pd.read_csv('Software_feature_shipment_data_set.csv')

# 1. Exploratory Data Analysis (EDA) Check for the Interviewer
print("Dataset Row/Column Shape:", df.shape)
print("\nMissing values per feature:\n", df.isnull().sum())

# 2. Data Cleaning: Impute missing records with the column median
numerical_cols = ['team_size', 'feature_complexity', 'num_dependencies', 'sprint_length_weeks',
                  'num_blockers', 'holidays_in_sprint', 'past_avg_delay_days', 'estimated_bug_count']

for col in numerical_cols:
    if df[col].isnull().sum() > 0:
        df[col] = df[col].fillna(df[col].median())

# 3. Feature Engineering: Compute Developer Load Pressure & Blocker Ratios
df['planned_shipment_date'] = pd.to_datetime(df['planned_shipment_date'])
df['complexity_per_developer'] = df['feature_complexity'] / (df['team_size'] + 1e-5)
df['blockers_per_dependency'] = df['num_blockers'] / (df['num_dependencies'] + 1e-5)

print("\n--- Preprocessing & Advanced Cleaning Complete ---")
```

```python
def generate_sprint_prompt(row):
    prompt = f"""### Instruction:
Analyze the engineering sprint metadata below and predict the number of final feature shipment delay days.

### Sprint Input Data:
Target Date: {row['planned_shipment_date'].strftime('%Y-%m-%d')}
Engineers Assigned: {row['team_size']}
Sprint Length: {row['sprint_length_weeks']} weeks
Complexity Score: {row['feature_complexity']:.2f}
Complexity/Dev Ratio: {row['complexity_per_developer']:.2f}
Total Blockers: {row['num_blockers']}
Historical Average Delays: {row['past_avg_delay_days']:.2f} days

### Response:
{int(row['delay_days'])} days<|endoftext|>"""
    return {"text": prompt}

# Convert tabular rows into structured sequence lists
text_dataset = [generate_sprint_prompt(row) for _, row in df.iterrows()]

# Structure directly into a Hugging Face Dataset object
from datasets import Dataset
hf_train_dataset = Dataset.from_list(text_dataset)

print(f"Successfully processed and formatted {len(hf_train_dataset)} examples for training!")
```

```python
from unsloth import FastLanguageModel
import torch
from trl import SFTTrainer
from transformers import TrainingArguments

# 1. Load the optimized 4-bit Phi-3 architecture
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/Phi-3-mini-4k-instruct",
    max_seq_length = 2048,
    dtype = None,
    load_in_4bit = True,
)

# 2. Inject Parameter-Efficient LoRA layers to capture tabular dynamics
model = FastLanguageModel.get_peft_model(
    model,
    r = 16,
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj", "gate_proj", "up_proj", "down_proj"],
    lora_alpha = 16,
    lora_dropout = 0,
    bias = "none",
    use_gradient_checkpointing = "unsloth",
    random_state = 42,
)

# 3. Setup the Supervised Fine-Tuning (SFT) Training Loop
trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = hf_train_dataset,
    dataset_text_field = "text",
    max_seq_length = 2048,
    dataset_num_proc = 2,
    packing = False,
    args = TrainingArguments(
        per_device_train_batch_size = 2,
        gradient_accumulation_steps = 4,
        warmup_steps = 5,
        max_steps = 60, # Safe threshold for quick validation
        learning_rate = 2e-4,
        fp16 = not torch.cuda.is_bf16_supported(),
        bf16 = torch.cuda.is_bf16_supported(),
        logging_steps = 1,
        output_dir = "outputs",
    ),
)

# Start training execution
trainer.train()
```

```python
import pandas as pd
import numpy as np
from datasets import Dataset
from unsloth import FastLanguageModel
import torch
from trl import SFTTrainer
from transformers import TrainingArguments

print("--- 1. Loading and Cleaning Dataset ---")
# Load the dataset
df = pd.read_csv('Software_feature_shipment_data_set.csv')

# Handle missing records safely using median metrics
numerical_cols = ['team_size', 'feature_complexity', 'num_dependencies', 'sprint_length_weeks',
                  'num_blockers', 'holidays_in_sprint', 'past_avg_delay_days', 'estimated_bug_count']
for col in numerical_cols:
    if df[col].isnull().sum() > 0:
        df[col] = df[col].fillna(df[col].median())

# Advanced feature engineering
df['planned_shipment_date'] = pd.to_datetime(df['planned_shipment_date'])
df['complexity_per_developer'] = df['feature_complexity'] / (df['team_size'] + 1e-5)
df['blockers_per_dependency'] = df['num_blockers'] / (df['num_dependencies'] + 1e-5)


print("--- 2. Mapping Tabular Data to Text Strings ---")
# Prompt generation mapping function
def generate_sprint_prompt(row):
    prompt = f"""### Instruction:
Analyze the engineering sprint metadata below and predict the number of final feature shipment delay days.

### Sprint Input Data:
Target Date: {row['planned_shipment_date'].strftime('%Y-%m-%d')}
Engineers Assigned: {row['team_size']}
Sprint Length: {row['sprint_length_weeks']} weeks
Complexity Score: {row['feature_complexity']:.2f}
Complexity/Dev Ratio: {row['complexity_per_developer']:.2f}
Total Blockers: {row['num_blockers']}
Historical Average Delays: {row['past_avg_delay_days']:.2f} days

### Response:
{int(row['delay_days'])} days<|endoftext|>"""
    return {"text": prompt}

# Package into a proper HuggingFace Dataset structure
text_dataset = [generate_sprint_prompt(row) for _, row in df.iterrows()]
hf_train_dataset = Dataset.from_list(text_dataset)


print("--- 3. Initializing Model and Starting Training Loop ---")
# Load model structure
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/Phi-3-mini-4k-instruct",
    max_seq_length = 2048,
    dtype = None,
    load_in_4bit = True,
)

# Inject Parameter-Efficient LoRA layers
model = FastLanguageModel.get_peft_model(
    model,
    r = 16,
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj", "gate_proj", "up_proj", "down_proj"],
    lora_alpha = 16,
    lora_dropout = 0,
    bias = "none",
    use_gradient_checkpointing = "unsloth",
    random_state = 42,
)

# Configure SFT Trainer with the fix to prevent Pickling Error
trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = hf_train_dataset,
    dataset_text_field = "text",
    max_seq_length = 2048,
    dataset_num_proc = 2,
    packing = False,
    args = TrainingArguments(
        per_device_train_batch_size = 2,
        gradient_accumulation_steps = 4,
        warmup_steps = 5,
        max_steps = 60,
        learning_rate = 2e-4,
        fp16 = not torch.cuda.is_bf16_supported(),
        bf16 = torch.cuda.is_bf16_supported(),
        logging_steps = 1,
        output_dir = "outputs",
        save_strategy = "no", # Prevents the pickling error crash at the end
    ),
)

# Execute Fine-tuning
trainer.train()
print("--- Training Successfully Finished! ---")
```
