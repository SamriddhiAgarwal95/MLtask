```python
import pandas as pd
import numpy as np

print("📖 Step 1: Loading and preprocessing dataset from Colab sidebar...")

# 1. Load the CSV file into a pandas dataframe
df = pd.read_csv('Software_feature_shipment_data_set.csv')

# 2. Handle missing data: Impute empty rows with the median value of that column
for col in df.select_dtypes(include=[np.number]).columns:
    df[col] = df[col].fillna(df[col].median())

# 3. Feature Engineering: Create the workload intensity ratio metric
if 'feature_complexity' in df.columns and 'team_size' in df.columns:
    df['complexity_per_developer'] = df['feature_complexity'] / df['team_size']
else:
    print("⚠️ Column mismatch. Current available columns are:")
    print(df.columns)

print("✅ Step 1 Complete: Data successfully parsed and engineered!")
print(f"📋 Dataset Size: {df.shape[0]} rows and {df.shape[1]} columns ready.")
```

```python
from sklearn.model_selection import train_test_split

print("📥 Step 2: Isolating evaluation and training environments...")

# Drop the text date and target column to isolate pure predictive numbers (X)
X = df.drop(columns=['delay_days', 'planned_shipment_date'], errors='ignore')
y = df['delay_days']

# Split data into 80% training rows and 20% validation testing rows
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

print("✅ Step 2 Complete: Data split successfully generated.")
print(f"📊 Training Partition: {X_train.shape[0]} samples (Used for learning patterns)")
print(f"📊 Testing Partition  : {X_test.shape[0]} samples (Kept completely hidden from models)")
```

```python
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb

print("⚙️ Step 3: Training traditional machine learning architectures...")

# 1. Instantiate and train Random Forest (Ensemble of independent decision trees)
rf_baseline = RandomForestRegressor(n_estimators=100, max_depth=8, random_state=42)
rf_baseline.fit(X_train, y_train)

# 2. Instantiate and train XGBoost (Sequential gradient boosting decision trees)
xgb_baseline = xgb.XGBRegressor(n_estimators=100, max_depth=6, learning_rate=0.1, random_state=42)
xgb_baseline.fit(X_train, y_train)

print("✅ Step 3 Complete: Both baseline models are successfully trained!")
```

```python
from sklearn.model_selection import GridSearchCV
import xgboost as xgb

print("🔧 Step 3.5: Fine-Tuning XGBoost Hyperparameters for MAE Optimization...")

# 1. Define the grid of math configurations to test
param_grid = {
    'max_depth': [4, 6, 8],
    'learning_rate': [0.05, 0.1, 0.2],
    'n_estimators': [50, 100, 150]
}

# 2. Initialize the tuning search engine focused on minimizing error
xgb_model = xgb.XGBRegressor(random_state=42)
grid_search = GridSearchCV(estimator=xgb_model, param_grid=param_grid,
                           scoring='neg_mean_absolute_error', cv=3, verbose=1)

# 3. Execute the cross-validation search on the 1,040 training rows
grid_search.fit(X_train, y_train)

# 4. Extract the absolute best fine-tuned settings
best_tuned_xgb = grid_search.best_estimator_

print("\n✅ Fine-Tuning Complete! Optimal configurations found and locked in.")
print(f"🎯 Best Parameters: {grid_search.best_params_}")
```

```python
from sklearn.metrics import mean_absolute_error

print("🔮 Step 4: Measuring predictive precision across the test partition...")

# Generate target predictions from your baseline and fine-tuned architectures
rf_preds = rf_baseline.predict(X_test)
xgb_preds = xgb_baseline.predict(X_test)
tuned_preds = best_tuned_xgb.predict(X_test)

# Calculate Mean Absolute Errors
rf_mae = mean_absolute_error(y_test, rf_preds)
xgb_mae = mean_absolute_error(y_test, xgb_preds)
tuned_mae = mean_absolute_error(y_test, tuned_preds)

print("\n================== 📊 SUBMISSION METRICS ==================")
print(f"🌲 Baseline Random Forest MAE  : {rf_mae:.3f} days")
print(f"⚡ Baseline XGBoost MAE        : {xgb_mae:.3f} days")
print(f"🔥 Fine-Tuned XGBoost MAE      : {tuned_mae:.3f} days")
print("===========================================================")

# Summary statement to prove impact
improvement = xgb_mae - tuned_mae
if improvement > 0:
    print(f"💡 Owner Insight: Hyperparameter tuning successfully optimized the model, reducing our timeline prediction error by {improvement:.2f} days!")
else:
    print("💡 Owner Insight: The baseline parameters were already highly optimal for this specific slice of historical data.")
```
